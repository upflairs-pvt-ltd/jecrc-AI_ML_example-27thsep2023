```python
plt.plot(df['total_bill'],color='r',marker='*' ,ms = 15,mfc='g',lw=1)
plt.title('line plot')
plt.xlabel('axis - x')
plt.ylabel('axis - y')
plt.show()
```

```python
plt.scatter(df['total_bill'],df['tip'],color='r',marker='*')
plt.title('scatter plot',size=20,color='b')
plt.xlabel('totoal bill',size=20,color='b')
plt.ylabel('tips',size=20,color='b')
plt.show()
```

```python
plt.bar(df['time'],df['total_bill'],ec='k',color = 'g')
plt.xlabel('time')
plt.ylabel('total bill')
plt.show()
```

```python
plt.bar(df['day'],df['total_bill'])
plt.xlabel('time')
plt.ylabel('total bill')
plt.show()
```

```python
plt.barh(df['day'],df['total_bill'])
plt.xlabel('time')
plt.ylabel('total bill')
plt.show()
```

```python
bins = [0,30,50,80]
plt.hist(df['total_bill'],bins,ec='k')
plt.show()
```

```python
plt.boxplot(df['total_bill'])
plt.show()
```

```python
label = list(df['day'].value_counts().index)
value = list(df['day'].value_counts().values)
```

```python

plt.pie(value,labels=label,explode=[0,0.3,0,0.2],autopct="%1.2f%%",shadow=True , wedgeprops={'ec':'k'})
plt.show()
```
